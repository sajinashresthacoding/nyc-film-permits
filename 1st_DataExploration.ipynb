```python
##import libraries
import pandas as pd
```

```python
df_permits = pd.read_csv("data/20240630_FilmPermits.csv")
df_permits
```

```python
##i can drop event agencies, community boards, entered on, and police precincts
##the start dates are not in order, i need to find a way to make them in order 
##i need to see how far the start dates go 
##i need to drop subcategories and categories (check notion to see what to drop)
```

```python
##double checking before dropping
df_permits['Country'].unique()
```

```python
df_permits['EventAgency'].unique()
```

```python
##dropping the not needed columns (agencies, community boards, entered on, police precincts)
df_permits = df_permits.drop(['EventAgency', 'CommunityBoard(s)', 'PolicePrecinct(s)'], axis=1)
df_permits
```

```python
##orderign the time because pandas doesnt understand time
df_permits['StartDateTime'] = df_permits['StartDateTime'].apply(pd.to_datetime)
df_permits
```

```python
df_permits['EndDateTime'] = df_permits['EndDateTime'].apply(pd.to_datetime)
df_permits
```

```python
##sort the dates of start date time 
df_OrderedDates = df_permits.sort_values(by = 'StartDateTime')
df_OrderedDates
```

```python
df_OrderedDates = df_OrderedDates.drop(['EnteredOn'], axis=1)
```

```python
df_OrderedDates
```

```python
##let's save this df into a csv, for keeping
df_OrderedDates.to_csv("data/File1_DatesInOrder.csv", index = False, encoding = "UTF-8")
```

```python
##check the uniques in category and subcategory
df_OrderedDates['EventType'].unique()
```

```python
##i dont know what a 'DCAS Prep/Shoot/Wrap Permit' is so i'll keep it
## i will keep: 'Shooting Permit'and 'DCAS Prep/Shoot/Wrap Permit'
```

```python
df_OrderedDates['Category'].unique()
```

```python
##lets do 'Television', 'Documentary', 'Film', 'Commercial','Music Video' for categories
##let's keep those
```

```python
df_OrderedDates['SubCategoryName'].unique()
```

```python
##let's do 'Episodic series','Cable-episodic', 'Short', 'Commercial', 'Feature','Made for TV/mini-series',
##'Signed Artist', 'Reality', 'Pilot','Independent Artist', 'Children', 'Trailers','Cable-daily'
```

```python
##FILTER1
##ok, let's start weeding EventType
##we are only keeping 'Shooting Permit'and 'DCAS Prep/Shoot/Wrap Permit'
df_filter1 = df_OrderedDates[df_OrderedDates['EventType'].isin\
                             (['Shooting Permit', 'DCAS Prep/Shoot/Wrap Permit'])]
df_filter1
```

```python
##check to see if your data is filtered correctly
df_filter1['EventType'].unique()
```

```python
##FILTER2
##filtering for category
##we are only keeping: 'Television', 'Documentary', 'Film', 'Commercial','Music Video'
df_filter2 = df_filter1[df_filter1['Category'].isin\
                             (['Television', 'Documentary', 'Film', 'Commercial','Music Video'])]
df_filter2
```

```python
##check to see if your data is filtered correctly
df_filter2['Category'].unique()
```

```python
##FILTER3
## for subcategory, we are keeping:
##'Episodic series','Cable-episodic', 'Short', 'Commercial', 'Feature','Made for TV/mini-series',
##'Signed Artist', 'Reality', 'Pilot','Independent Artist', 'Children', 'Trailers','Cable-daily'
df_filter3 = df_filter2[df_filter2['SubCategoryName'].isin\
                             (['Episodic series','Cable-episodic', 'Short', 'Commercial',\
                               'Feature','Made for TV/mini-series','Signed Artist', 'Reality', 'Pilot',\
                              'Independent Artist', 'Children', 'Trailers','Cable-daily'])]
df_filter3
```

```python
##check to see if your data is filtered correctly
df_filter3['SubCategoryName'].unique()
```

```python
##let's save this data
df_filter3.to_csv("data/File2_FilteredData.csv", index = False, encoding = "UTF-8")
```
