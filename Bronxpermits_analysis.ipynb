```python
#import libraries
import pandas as pd
```

```python
df_bx = pd.read_csv("data/BronxData.csv")
df_bx
```

```python
#finding out how many different things were shot in the city
type_bx = df_bx.groupby("Category")
type_bx
```

```python
categories = type_bx["ZipCode(s)"].count()
categories
```

```python
##bronx zipcodes
bx_zipcodes = [ 
    10467,
    10456,
    10458,
    10453,
    10468,
    10457,
    10452,
    10469,
    10462,
    10466,
    10463,
    10472,
    10460,
    10473,
    10451,
    10461,
    10459,
    10465,
    10475,
    10455,
    10454,
    10471,
    10474,
    10470,
    10803,
    10464,
    10499 ]
```

```python
##now to run this list of zipcodes to the dataframe and get counts
counts = {}
for bx_zipcodes in df_bx['ZipCode(s)']:
    counts[bx_zipcodes] = counts.get(bx_zipcodes, 0) + 1
print(counts)
```

```python
df_zipcodes = pd.DataFrame.from_dict(counts, orient='index')
df_zipcodes
```

```python
ordered = df_zipcodes.sort_values(by=0, ascending=False)
ordered
```

```python
ordered.to_csv("data/BronxZipcodesCount.csv", index = True, encoding = "UTF-8")
```
