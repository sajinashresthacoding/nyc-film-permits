```python
##now to run this lizt of zipcodes to the dataframe and get counts
counts = {}
for nyc_zipcodes in df_nyc['ZipCode(s)']:
    counts[nyc_zipcodes] = counts.get(nyc_zipcodes, 0) + 1
print(counts)
```

```python
df_zipcodes = pd.DataFrame.from_dict(counts, orient='index')
df_zipcodes
```

```python
ordered = df_zipcodes.sort_values(by=0, ascending=False)
```
