```python
##import libraries
import pandas as pd
```

```python
df = pd.read_csv("data/File6_BronxStatenIslandData.csv")
df
```

```python
df_si = df[df['Borough'].isin(['Staten Island'])]
```

```python
df_si.reset_index(drop=True, inplace=True)
df_si
```

```python
##saveSI
df_si.to_csv("data/StatenIslandData.csv", index = False, encoding = "UTF-8")
```

```python
df_bx.to_csv("data/BronxData.csv", index = False, encoding = "UTF-8")
```

```python
type_si = df_si.groupby("Category")
type_si
```

```python
categories = type_si["ZipCode(s)"].count()
categories
```

```python
si_zipcodes = [10301, 10302, 10303, 10304, 10305, 10306, 10307, 10308, 10309, 10310, 10311, 10312, 10313, 10314]
si_zipcodes
```

```python
##now to run this list of zipcodes to the dataframe and get counts
counts = {}
for si_zipcodes in df_si['ZipCode(s)']:
    counts[si_zipcodes] = counts.get(si_zipcodes, 0) + 1
print(counts)
```

```python
df_zipcodes = pd.DataFrame.from_dict(counts, orient='index')
df_zipcodes
```

```python
ordered = df_zipcodes.sort_values(by=0, ascending=False)
ordered
```

```python
ordered.to_csv("data/SIZipcodesCount.csv", index = True, encoding = "UTF-8")
```
